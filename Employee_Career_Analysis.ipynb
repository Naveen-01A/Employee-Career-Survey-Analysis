```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/content/Career Aspirations of GenZ.csv')
```

```python
df.head()
```

```python
df.columns
```

### 1) What is the analysis of country?


```python
country = df['Your Current Country.'].value_counts()
```

```python
country
```

```python
label = country.index
counts = country.values
colors = ['red', 'lightgreen']
fig = go.Figure(data = [go.Pie(labels = label, values = counts)])
fig.update_layout(title_text = 'Current country')
fig.update_traces(hoverinfo = 'label+value', textinfo = 'percent', textfont_size = 20,
                  marker = dict(colors = colors, line = dict(color = 'black', width = 3)))

fig.show()
```

Here , we use pie chart to analysis data as per the country wise percentage.

### 2) What are the factors influencing the carrer aspirations of Genz (youth or youngsters)?

```python
question1 = df['Which of the below factors influence the most about your career aspirations ?']
```

```python
question1.unique()
```

```python
question= df['Which of the below factors influence the most about your career aspirations ?'].value_counts()
label = question.index
counts = question.values
colors = ['gold', 'lightgreen']
fig = go.Figure(data=[go.Pie(labels = label, values = counts)])
fig.update_layout(title_text = 'Factors influencing career aspirations')
fig.update_traces(hoverinfo = 'label+value', textinfo = 'percent', textfont_size = 20,
                  marker = dict(colors = colors, line = dict(color='black', width = 3)))

fig.show()
```
